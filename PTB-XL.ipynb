```python
import pandas as pd

# Load the main PTB-XL metadata file
df = pd.read_csv('ptbxl_database.csv')

# See the first few rows
df.head()
```

```python
# Show all unique values in the 'nurse' column
unique_nurses = df['nurse'].unique()
print("Distinct values in 'nurse' column:")
print(unique_nurses)
```

```python
# Show all unique values in the column
unique_values = df['scp_codes'].unique()
print(unique_values)
```

```python
# Count of each unique value in the column
unique_counts = df['scp_codes'].value_counts()
print("Frequency:")
print(unique_counts)
```

```python
# Drop unnecessary columns to reduce size
df_reduced = df.drop(columns=[
    'validated_by_human',
    'initial_autogenerated_report',
    'pacemaker',
    'filename_lr',
    'filename_hr',
    'baseline_drift',
    'static_noise',
    'burst_noise',
    'electrodes_problems'
])
```

```python
print(df_reduced.columns)
```

```python
# Print number of columns
print(f"Number of columns: {len(df_reduced.columns)}")
```

```python
import numpy as np

# Convert the nurse, site, validated_by columns

df_reduced['nurse'] = df_reduced['nurse'].apply(lambda x: f'Nurse{int(x)}' if pd.notna(x) else 'Unknown')
df_reduced['site'] = df_reduced['site'].apply(lambda x: f'Site{int(x)}' if pd.notna(x) else 'Unknown')
df_reduced['validated_by'] = df_reduced['validated_by'].apply(lambda x: f'Validator{int(x)}' if pd.notna(x) else 'Unknown')
```

```python
df_reduced['age'] = df_reduced['age'].astype('Int64')  # preserves NaNs
df_reduced['patient_id'] = df_reduced['patient_id'].astype('Int64')  # preserves NaNs
df_reduced['height'] = df_reduced['height'].astype('Int64')  # preserves NaNs
df_reduced['weight'] = df_reduced['weight'].astype('Int64')  # preserves NaNs
```

```python
df_reduced['recording_date'] = pd.to_datetime(df_reduced['recording_date'], utc=True)
```

```python
df_reduced['sex'] = df_reduced['sex'].astype(int).map({0: 'male', 1: 'female'})
```
